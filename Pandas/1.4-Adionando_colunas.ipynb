FUN√á√ïES E LAMBDA EM ASSIGN 

```python
#Utilizando uma fun√ß√£o 
def funcao_multiplicacao(df) : 
    return df.area * df.rooms
```

```python
base_copia.assign(multiplicacao = funcao_multiplicacao).loc[:5, ["area", "rooms", "multiplicacao"]]
```

```python
#Utilizando um lambda
base_copia.assign(soma_lambda = lambda x: x.rooms + x.bathroom,multiplicacao = funcao_multiplicacao).loc[:5,:]
```
